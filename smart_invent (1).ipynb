```python
data_cp = all_data.copy()
data_cp.head()
```

```python
#LABEL ENCODING
from sklearn.preprocessing import LabelEncoder

lb_enc = LabelEncoder()
data_cp["make_Cent_type"] = lb_enc.fit_transform(data_cp["center_type"])
data_cp[["center_type", "make_Cent_type"]].head(10)
```

```python
data_cp["make_category"] = lb_enc.fit_transform(data_cp["category"])
data_cp[["category", "make_category"]].head(10)
```

```python
data_cp["make_cuisine"] = lb_enc.fit_transform(data_cp["cuisine"])
data_cp[["cuisine", "make_cuisine"]].head(10)
```

```python
cp_data = data_cp.drop(['center_type','category','cuisine'], axis=1)
cp_data.head()
```

```python
##Traning and Testing data spliting
X = cp_data.drop('num_orders', axis=1)


y = cp_data['num_orders']
```

```python
X.shape
```

```python
X = X.to_numpy().reshape((X.shape[0], X.shape[1], 1))
```

```python
y.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
# Split data into training and test sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Check the shape
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# Define LSTM model
model = Sequential()

model.add(LSTM(units=50, return_sequences=False, input_shape=(X_train.shape[1], X_train.shape[2])))


# Add a Dense layer (output layer)
model.add(Dense(units=1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
# Train the LSTM model
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test))
```

```python
# Evaluate the model on test data
test_loss = model.evaluate(X_test, y_test)
print(f'Test Loss: {test_loss}')
```

```python
# Predict using the trained model
y_pred = model.predict(X_test)

# Plot predictions vs actual values
plt.figure(figsize=(12, 6))
plt.plot(y_test, label='True Values', color='blue')       # Blue for true
plt.plot(y_pred, label='Predicted Values', color='orange') # Orange for predicted
plt.title('LSTM Predictions vs True Values')
plt.xlabel('Time Step')
plt.ylabel('Normalized Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Example: Constants
unit_price = 1000  # price per product (m)
fixed_cost = 100000  # constant cost (C)

# Apply financial equation: y = m * x(t) - C
# where x(t) = predicted sales from LSTM (y_pred)
financials = (unit_price * y_pred) - fixed_cost

# Plot financials
plt.figure(figsize=(10, 5))
plt.plot(financials, label='Predicted Profit', color='green')
plt.title('Forecasted Financial Profit Over Time')
plt.xlabel('Time')
plt.ylabel('Profit (in )')
plt.legend()
plt.grid(True)
plt.show()
```
